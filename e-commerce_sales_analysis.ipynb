```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Amazon Sale Report.csv')
print(data)
```

```python
data.dtypes
```

```python
data.info()
```

```python
# Changing datatypes
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['ship-postal-code'] = data['ship-postal-code'].astype('str')
```

```python
data['Courier Status'].value_counts()
```

```python
# Checking missing values
data.isnull().sum()
```

```python
data['promotion-ids'] = data['promotion-ids'].fillna('No Promo')
```

```python
data.drop(columns=['Unnamed: 22'], inplace=True)

```

```python
data.drop(columns=['fulfilled-by'], inplace=True)
```

```python
data.drop(columns=['ship-city'], inplace=True)
```

```python
data.drop(columns=['currency'],inplace=True)
```

```python
data.drop(columns=['ship-country'], inplace=True)
```

```python
data = data.dropna(subset=['Amount'])
data['Courier Status'] = data['Courier Status'].fillna('Unknown')
```

```python
data = data.dropna(subset=['ship-state'])
```

```python
data.isnull().sum()
```

```python
# Checking Duplicates
data.duplicated().sum()
```

```python
data.duplicated(subset=['Order ID', 'SKU']).sum()
data[data.duplicated(subset=['Order ID', 'SKU'], keep=False)]

```

```python
data = data.drop_duplicates(subset=['Order ID', 'SKU'])
```

```python
# Overall Business Performance
total_sales = data['Amount'].sum()
total_orders = data['Order ID'].nunique()
total_qty = data['Qty'].sum()

print(total_sales, total_orders, total_qty)
```

```python
# Time Series Analysis
monthly_sales = data.groupby(data['Date'].dt.to_period('M'))['Amount'].sum()
monthly_sales
```

```python
# Category Performance
category_sales = data.groupby('Category')['Amount'].sum().sort_values(ascending=False)
print(category_sales)
category_percent = (category_sales / category_sales.sum()) * 100
category_percent = category_percent.round(2)
print(category_percent)
```

```python
# Product Analysis
top_products = data.groupby('SKU')['Amount'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# Location Analysis
# Top 5 state with highest sales
State_sales = data.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(5)
State_sales
```

```python
# Order status
status_counts = data['Status'].value_counts()
print(status_counts)
status_percent = (status_counts/status_counts.sum())*100
status_percent = status_percent.round(2)
print(status_percent)
```

```python
# Courier Status
data['Courier Status'].value_counts()
```

```python
# Converting to csv
data.to_csv("amazon_sales_cleaned.csv", index=False)
```
